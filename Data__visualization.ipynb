```python
fig = plt.figure(figsize = (15,3))
plt.plot(vis_data['month_number'].tolist(),vis_data['total_profit'].tolist())
plt.xlabel("month_number")
plt.ylabel("total_profit")
plt.title("Sales Graph")
#fig.savefig('sales graph.png', dpi=300, bbox_inches='tight')
```

```python
plt.subplots(1,1)        #Create a figure and a set of subplots.
                          # 1 row 1 column
```

```python
plt.subplots(2, 4)                                            # 2 rows n 4 columns
```

```python
plt.subplot(221)                 # the figure has 2 row, 2 columns, and this plot is the first plot.
# equivalent but more general
ax1 = plt.subplot(2, 2, 1)

# add a subplot with no frame     #the figure has 2 row, 2 columns, and this plot is the second plot.
ax2 = plt.subplot(222, frameon=False)

# add a polar subplot
plt.subplot(223, projection='polar')

# add a red subplot that shares the x-axis with ax1
plt.subplot(224, sharex=ax1, facecolor='red')

# delete ax2 from the figure
plt.delaxes(ax2)

# add ax2 to the figure again
plt.subplot(ax2)

# make the first axes "current" again
plt.subplot(221)
```

## Seaborn

```python
#strip plot

sns.stripplot(y= 'bathingsoap', data=vis_data,jitter=True) 
plt.ylabel('total units')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
sns.stripplot(x = 'month_number', y= 'bathingsoap', data=vis_data,jitter=True) 
plt.ylabel('bathingsoap')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
sns.stripplot(x = 'month_number', y= 'bathingsoap', data=vis_data, size = 20, jitter=False) 
plt.ylabel('bathingsoap')
plt.show()
```

```python
#Swarm plot

sns.swarmplot(x='toothpaste', y='month_number', data=vis_data, orient = 'h') #More no of people has given the tips
```

### Box and Violin plot

```python
sns.boxplot(x='month_number', y='total_units', data=vis_data)
plt.ylabel('Total Units')
```

### Bivariate

```python
#Joint plot

sns.jointplot(x= 'month_number', y= 'total_units', data=vis_data) # Most of the tips are between 2 to 4 if the tot bill is between 10 to 20.
plt.show()
```

```python
# Density plot
sns.jointplot(x= 'month_number', y= 'total_units', data=vis_data,kind='kde')
plt.show()
```

```python
#Pair plot

sns.pairplot(vis_data)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.pairplot(vis_data, hue='month_number') #color coding
plt.show()
```
